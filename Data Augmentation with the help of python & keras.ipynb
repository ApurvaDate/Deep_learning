```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
#import imageDatagenerator which helps in applying data augmentation properties.
```

```python
datagen = ImageDataGenerator(
                rotation_range=40,
                width_shift_range=0.2,
                height_shift_range=0.2,
                shear_range=0.2,
                zoom_range=0.2,
                horizontal_flip=True,
                fill_mode="nearest")
```

```python
%%time
img = load_img("./cat.jpg")
x = img_to_array(img) #this is a numpy array
x = x.reshape((1,) + x.shape) #reshape the array

#the .flow() command below  generates batches of randomly transformed images
#and the saves the results to the "preview/" directory.
i = 0
for batch in datagen.flow(x, batch_size = 1, save_to_dir="preview", save_prefix="cat", save_format="jpeg"):
    i += 1
    if i > 20:  #we save only 20 images
        break #otherwise the generator would loop indefinitely.
```
